[机器学习--Logistic回归计算过程的推导](https://blog.csdn.net/ligang_csdn/article/details/53838743)

```
import logRegres
```

```
dataArr,labelMat=logRegres.loadDataSet()
weights=logRegres.gradAscent(dataArr,labelMat)
```

```
# 得到的权值为矩阵
weights
```

```
#获取数组
weights.getA()
```

此处的线性模型为：
$label = \omega_0 + \omega_1 x_1 +\omega_2 x_2$

令$label=0$则得到三维空间的一个平面用于区分不同类别。将该三维空间投影到二维平面（便于画图），则有：
$$
0=\omega_0 + \omega_1 x_1 +\omega_2 x_2
$$

$$
x_2 = \frac {-\omega_0 - \omega_1 x_1}{\omega_2}
$$
 
 对应`plotBestFit()`中第50行

```
logRegres.plotBestFit(weights.getA())
```
